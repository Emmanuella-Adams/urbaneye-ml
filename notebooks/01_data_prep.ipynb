```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load the dataset
data = pd.read_csv('sample_data/dataset.csv')  # Update with actual dataset path

# Display the first few rows of the dataset
data.head()
```

```python
# Data cleaning
# Example: Remove duplicates
data = data.drop_duplicates()

# Handle missing values
data.fillna(method='ffill', inplace=True)  # Forward fill for missing values

# Display cleaned data
data.info()
```

```python
# Data transformation
# Example: Convert categorical variables to numerical
data['category'] = data['category'].astype('category').cat.codes

# Display transformed data
data.head()
```

```python
# Save the cleaned and transformed data
data.to_csv('sample_data/cleaned_data.csv', index=False)  # Update with actual path

# End of data preparation
```
